```python
import json
import numpy as np
import matplotlib.pyplot as plt


def txt_read(path):
    data_first_params = []
    data_last_params = []
    with open(path, "r") as f:
        for line in f:
            try:
                data_txt = line.split('||')
                data_first_params.append(eval(data_txt[6].strip().split(":")[-1]))
                data_last_params.append(eval(data_txt[7].strip().split(":")[-1]))
            except:
                ...
            
    return data_first_params, data_last_params
def main(path, number: int = -1, last=True):
    # plt.rcParams["font.family"] = "Times New Roman"
    data_first, data_last = txt_read(path)
    data_first, data_last = np.array(data_first[:number]), np.array(data_last[:number])
    param_length = len(data_first[0])

    
    labels = [
        "LoRA_r2",
        "LoRA_r4",
        "LoRA_r8",
        "LoRA_gelu_r2",
        "LoRA_gelu_r4",
        "LoRA_gelu_r8",
        "LoRA_decom_r2",
        "LoRA_decom_r4",
        "LoRA_decom_r8",
    ]
    steps = [_ for _ in range(len(data_first))]
    # labels_last
    if last:
        plt.figure(figsize=(8,4))
        for _ in range(len(labels)):
            plt.plot(steps, data_last[:, _], label=labels[_])
    else:
        plt.figure(figsize=(8,4))
        for _ in range(len(labels)):
            plt.plot(steps, data_first[:, _], label=labels[_])
    
    # fig.tight_layout()
    plt.xlabel('Step')
    plt.ylabel('Weight Value')
    plt.legend(loc=2, fontsize=6)
    plt.grid(axis='y')
    ax = plt.gca()
    y_color = (0.3, 0.3, 0.3)
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    # ax.spines['left'].set_visible(False)
    ax.spines['left'].set_color(y_color)
    ax.spines['bottom'].set_visible(False)
    # ax.axhline(y=-1.2,linewidth=3, color=(0.96862745, 0.76078431, 0.71764706))
    # ax.axhline(y=-1.6,linewidth=1, color=y_color)
    ax.axhline(y=-1.2, linewidth=2, color=y_color)
    # ax.axvline(linewidth=2, color="green")
    # plt.show()p
    # ax.get_yticklabels()[3].set_color((0.96862745, 0.76078431, 0.71764706))
    plt.tick_params(axis='y', colors=y_color)
    plt.tick_params(axis='x', colors=y_color)
    plt.savefig("/mnt/bn/ecom-ccr-dev/mlx/users/dongjunwei/EasyGuard/examples/peft_llm/logs/weight_change_first.pdf", format="pdf", bbox_inches="tight")

path = "/mnt/bn/ecom-ccr-dev/mlx/users/dongjunwei/EasyGuard/examples/peft_llm/model_output/naslora_v4/llama2/boolq/independent/peft_naslora_lr1e-3_boolq_independent_no_mlc_arch_lr01_pre_combibe_70_10_steplr_10/trainer.log"
main(path, last=False)

```
